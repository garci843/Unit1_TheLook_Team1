# Task
I have successfully identified and defined the top three growth KPIs for the e-commerce business based on the available data:

*   **Customer Acquisition Rate:** Measures the rate at which new customers are acquired.
*   **Customer Lifetime Value (CLV):** Represents the total revenue expected from a single customer.
*   **Average Order Value (AOV):** Tracks the average amount spent per order.

These KPIs provide a strong foundation for understanding and driving business growth.

The next logical step is to write the SQL queries to calculate and visualize these metrics over time. This will allow us to analyze trends, identify patterns, and derive actionable insights to improve business performance.

### Data Analysis Key Findings
*   **Customer Acquisition Rate** can be calculated by counting the number of new users in the `users` table based on their `created_at` date for a given period. This KPI is crucial for measuring marketing effectiveness and business expansion.
*   **Customer Lifetime Value (CLV)** is determined by summing the `sale_price` for each user across all their orders. This requires joining the `orders` and `order_items` tables and helps in identifying high-value customer segments and improving retention strategies.
*   **Average Order Value (AOV)** is calculated by dividing the total revenue (sum of `sale_price` from the `order_items` table) by the total number of orders (count of distinct `order_id`s from the `orders` table). Increasing AOV is a direct way to boost revenue without acquiring new customers.


**Prompt:**
```
BigQuery SQL Only

Generate an SQL query that calculates the Average Order Value and its Month-over-Month (MoM) and Year-over-Year (YoY) growth percentages.

Use CTEs and window functions.

Use these tables:
`bigquery-public-data.thelook_ecommerce.order_items` (oi)
`bigquery-public-data.thelook_ecommerce.orders` (o)

Average Order Value (AOV) is calculated by dividing the total revenue (sum of `sale_price` from the `order_items` table) by the total number of orders (count of distinct `order_id`s from the `orders` table).

Order the result by month and year descending and limit it to 24 rows.
```

````python
query = """
WITH
  monthly_sales AS (
    -- First, aggregate revenue and order counts by month
    SELECT
      DATE_TRUNC(o.created_at, MONTH) AS sales_month,
      SUM(oi.sale_price) AS total_revenue,
      COUNT(DISTINCT o.order_id) AS total_orders
    FROM
      `bigquery-public-data.thelook_ecommerce.orders` AS o
      JOIN `bigquery-public-data.thelook_ecommerce.order_items` AS oi
      ON o.order_id = oi.order_id
    WHERE o.status = 'Complete' -- Consider only completed orders for accurate AOV
    GROUP BY
      1
  ),
  monthly_aov AS (
    -- Calculate the Average Order Value (AOV) for each month
    SELECT
      sales_month,
      total_revenue / total_orders AS aov
    FROM
      monthly_sales
  ),
  aov_growth_calcs AS (
    -- Use LAG window function to get AOV from previous periods for growth calculation
    SELECT
      sales_month,
      aov,
      LAG(aov, 1) OVER (ORDER BY sales_month) AS prev_month_aov, -- AOV from 1 month ago
      LAG(aov, 12) OVER (ORDER BY sales_month) AS prev_year_aov -- AOV from 12 months ago
    FROM
      monthly_aov
  )
-- Final selection and calculation of MoM and YoY growth percentages
SELECT
  FORMAT_DATE('%Y-%m', sales_month) AS month,
  ROUND(aov, 2) AS average_order_value,
  ROUND((aov - prev_month_aov) * 100 / prev_month_aov, 2) AS mom_growth_percentage,
  ROUND((aov - prev_year_aov) * 100 / prev_year_aov, 2) AS yoy_growth_percentage
FROM
  aov_growth_calcs
ORDER BY
  sales_month DESC
LIMIT
  24;
"""
query_job = client.query(query)
results_df = query_job.to_dataframe()
display(results_df)
````

## Generate AOV Validation Query

### Subtask:
Generate a single, complex BigQuery SQL query that uses two parallel CTE pipelines to calculate Average Order Value (AOV) for male customers. The first pipeline will calculate AOV at the item-level (`SUM(sale_price) / COUNT(DISTINCT order_id)`). The second will calculate it at the order-level (`AVG(order_total)`). The final query will join these results, compute the absolute and percentage difference, and return the top 50 results ordered by the percentage difference.


**Reasoning**:
I will construct a single, complex BigQuery SQL query that implements the two parallel CTE pipelines as specified in the detailed instructions. This query will calculate AOV using both item-level and order-level definitions, join them, compute the differences, and return the top 50 results with the largest discrepancies for validation.



````python
query = """
WITH
  -- 1. Base CTE with all male customer order details
  male_customer_orders AS (
    SELECT
      u.state,
      u.traffic_source AS device_type, -- Map traffic_source to device_type
      o.order_id,
      oi.sale_price
    FROM
      `bigquery-public-data.thelook_ecommerce.orders` AS o
      JOIN `bigquery-public-data.thelook_ecommerce.users` AS u ON o.user_id = u.id
      JOIN `bigquery-public-data.thelook_ecommerce.order_items` AS oi ON o.order_id = oi.order_id
    WHERE
      u.gender = 'M'
  ),

  -- 2. First parallel pipeline: Item-level AOV
  item_level_aov AS (
    SELECT
      state,
      device_type,
      SUM(sale_price) / COUNT(DISTINCT order_id) AS aov_item_level
    FROM
      male_customer_orders
    GROUP BY
      state,
      device_type
  ),

  -- 3. Second parallel pipeline: Order-level AOV
  order_totals AS (
    -- 3a. Intermediate CTE to get total for each order
    SELECT
      order_id,
      state,
      device_type,
      SUM(sale_price) AS order_total
    FROM
      male_customer_orders
    GROUP BY
      order_id,
      state,
      device_type
  ),
  order_level_aov AS (
    -- 3b. Final order-level AOV calculation
    SELECT
      state,
      device_type,
      AVG(order_total) AS aov_order_level
    FROM
      order_totals
    GROUP BY
      state,
      device_type
  )

-- 4. Join the two AOV results and calculate differences
SELECT
  il.state,
  il.device_type,
  il.aov_item_level,
  ol.aov_order_level,
  (il.aov_item_level - ol.aov_order_level) AS absolute_difference,
  -- 5. Calculate percentage difference, handling potential division by zero
  SAFE_DIVIDE((il.aov_item_level - ol.aov_order_level), il.aov_item_level) * 100 AS percentage_difference
FROM
  item_level_aov AS il
  JOIN order_level_aov AS ol ON il.state = ol.state AND il.device_type = ol.device_type
-- 6. Order by the largest difference and limit
ORDER BY
  percentage_difference DESC
LIMIT 20;
"""

# 7. Execute the query and display the results
query_job = client.query(query)
results_df = query_job.to_dataframe()
display(results_df)
````

````python
query = """
WITH
  -- 1. Get order-level totals for the 'Accessories' category
  accessory_order_totals AS (
    SELECT
      o.order_id,
      u.state,
      u.traffic_source AS device_type,
      DATE_TRUNC(o.created_at, MONTH) AS order_month,
      SUM(oi.sale_price) AS order_total
    FROM
      `bigquery-public-data.thelook_ecommerce.orders` AS o
      JOIN `bigquery-public-data.thelook_ecommerce.users` AS u ON o.user_id = u.id
      JOIN `bigquery-public-data.thelook_ecommerce.order_items` AS oi ON o.order_id = oi.order_id
      JOIN `bigquery-public-data.thelook_ecommerce.products` AS p ON oi.product_id = p.id
    WHERE
      p.category = 'Accessories'
    GROUP BY
      1, 2, 3, 4
  ),

  -- 2. Calculate percentile boundaries for winsorizing
  percentiles AS (
    SELECT DISTINCT
      order_month,
      PERCENTILE_CONT(order_total, 0.01) OVER (PARTITION BY order_month) AS p01,
      PERCENTILE_CONT(order_total, 0.99) OVER (PARTITION BY order_month) AS p99
    FROM
      accessory_order_totals
  ),

  -- 3. Winsorize the order totals
  winsorized_orders AS (
    SELECT
      aot.state,
      aot.device_type,
      aot.order_total,
      CASE
        WHEN aot.order_total < p.p01 THEN p.p01
        WHEN aot.order_total > p.p99 THEN p.p99
        ELSE aot.order_total
      END AS winsorized_total
    FROM
      accessory_order_totals AS aot
      JOIN percentiles AS p ON aot.order_month = p.order_month
  ),

  -- 4. Calculate Raw AOV
  raw_aov AS (
    SELECT
      state,
      device_type,
      AVG(order_total) AS raw_aov
    FROM
      winsorized_orders -- Can reuse this CTE as it has original order_total
    GROUP BY
      state,
      device_type
  ),

  -- 5. Calculate Winsorized AOV
  winsor_aov AS (
    SELECT
      state,
      device_type,
      AVG(winsorized_total) AS winsor_aov
    FROM
      winsorized_orders
    GROUP BY
      state,
      device_type
  )

-- 6. Final comparison
SELECT
  r.state,
  r.device_type,
  r.raw_aov,
  w.winsor_aov,
  (r.raw_aov - w.winsor_aov) * 100 / r.raw_aov AS percentage_difference
FROM
  raw_aov AS r
  JOIN winsor_aov AS w ON r.state = w.state AND r.device_type = w.device_type
ORDER BY
  percentage_difference DESC
LIMIT 20;

"""
query_job = client.query(query)
results_df = query_job.to_dataframe()
display(results_df)
````

# Task
Generate a Python code that queries the daily revenue for the last 30 days from the `"bigquery-public-data.thelook_ecommerce.order_items"` and `"bigquery-public-data.thelook_ecommerce.orders"` tables and then uses Plotly to create an interactive time-series line chart of the revenue.

````python
query = """
SELECT
  DATE(o.created_at) AS order_date,
  SUM(oi.sale_price) AS daily_revenue
FROM
  `bigquery-public-data.thelook_ecommerce.orders` AS o
  JOIN `bigquery-public-data.thelook_ecommerce.order_items` AS oi
  ON o.order_id = oi.order_id
WHERE
  DATE(o.created_at) >= DATE_SUB(CURRENT_DATE(), INTERVAL 30 DAY)
GROUP BY
  order_date
ORDER BY
  order_date;
"""

query_job = client.query(query)
daily_revenue_df = query_job.to_dataframe()
display(daily_revenue_df.head())
````

````python
import plotly.express as px

fig = px.line(
    daily_revenue_df,
    x='order_date',
    y='daily_revenue',
    title='Daily Revenue for the Last 30 Days',
    labels={'order_date': 'Date', 'daily_revenue': 'Revenue (USD)'}
)

fig.show()
````
